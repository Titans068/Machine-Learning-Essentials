```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("ram_price.csv")
df
```

```python
plt.semilogy(df.date, df.price)
plt.xlabel("Year")
plt.ylabel("Price in $/Mbyte")
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
max(df.date)
```

```python
train = df.loc[df.date < 2000]
test = df.loc[df.date >= 2000]
x_train = np.array(train.date)[:, np.newaxis].copy()
y_train = train.price.copy()
display(x_train)
display(y_train)
```

```python
tree = DecisionTreeRegressor()
tree
```

```python
tree.fit(x_train, y_train)
```

```python
x_all = df.date[:, np.newaxis].copy()
x_all
```

```python
pred = tree.predict(x_all)
pred
```

```python
plt.semilogy(train.date, train.price, label="Training data")
plt.semilogy(test.date, test.price, label="Training data")
plt.semilogy(df.date, pred, label="Tree prediction")
plt.legend()
```
