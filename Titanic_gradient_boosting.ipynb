```python
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
encoder
```

```python
print(type(titanic))
titanic.columns
```

```python
#sex, embarked, class, who, adult_male,deck,embark_town,alive,alone
titanic.sex=encoder.fit_transform(titanic.sex)
titanic.embarked=encoder.fit_transform(titanic.embarked)
titanic["class"]=encoder.fit_transform(titanic["class"])
titanic.who=encoder.fit_transform(titanic.who)
titanic.adult_male=encoder.fit_transform(titanic.adult_male)
titanic.deck=encoder.fit_transform(titanic.deck)
titanic.embark_town=encoder.fit_transform(titanic.embark_town)
titanic.alive=encoder.fit_transform(titanic.alive)
titanic.alone=encoder.fit_transform(titanic.alone)
titanic
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imputer=SimpleImputer(missing_values=np.nan)
imputer
```

```python
titanic.age=imputer.fit_transform(titanic[["age"]])
titanic
```

```python
x=titanic.drop("survived",axis=1)
y=titanic.survived.copy()
display(x)
display(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
display(x_train)
display(x_test)
display(y_train)
display(y_test)
```

```python
len(titanic)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf=GradientBoostingClassifier(max_depth=4, min_samples_split=35, min_samples_leaf=45, subsample=.7, learning_rate=.06)
clf
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x,y)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
pd.DataFrame({"actual":y_test,"predicted":y_pred})
```

```python
from sklearn.metrics import r2_score, confusion_matrix, roc_auc_score, roc_curve, classification_report
```

```python
r2_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d',cmap="Reds")
```

```python
y_proba=clf.predict_proba(x_test)
y_proba
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_proba[:,1])
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

```python
roc_auc_score(y_test,y_proba[:,1])
```
