```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
cancer = load_breast_cancer()
cancer
```

```python
cancer_df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
cancer_df
```

```python
print("Shape of cancer data is {}".format(cancer_df.shape))
```

```python
cancer_df.describe()
```

```python
x = cancer.data
y = cancer.target
display(x)
display(y)
```

```python
len(x), len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=1)
display(x_train)
display(y_test)
display(y_train)
display(y_test)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=0)
tree_model
```

```python
tree_model.fit(x_train, y_train)
```

```python
print("Accuracy on training set is %.3f" % tree_model.score(x_train, y_train))
print("Accuracy on test set is %.3f" % tree_model.score(x_test, y_test))
print("Accuracy is %.3f" % tree_model.score(x, y))
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(tree_model, out_file="graph.dot", class_names=["malignant", "benign"],
                feature_names=cancer.feature_names, impurity=False, filled=True)
```

```python
import numpy as np
```

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")


plot_feature_importances_cancer(tree_model)
```
