```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([5,10,15,20,25]).reshape(-1, 1)
y = np.array([6,8,7,12,15])

model = LinearRegression()
model.fit(x,y)
display(model.coef_)
display(model.intercept_)

plt.scatter(x, y, marker='o', color='orchid')
plt.plot(x, (x * model.coef_ + model.intercept_), color='slateblue')

x_new = np.arange(0, 10).reshape(-1,1)
print(x_new)

y_new = model.predict(x_new)
print(len(y_new))
y_new
```

```python
plt.scatter(y_new, (model.intercept_ * y_new) + model.coef_)
plt.plot(y_new, (model.intercept_ * y_new) + model.coef_)
```

```python
plt.scatter(x_new, (y_new * model.intercept_) + model.coef_)
plt.plot(x_new, (y_new * model.intercept_) + model.coef_)
```
